```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.genfromtxt('ex3x.dat')
x
y = np.genfromtxt('ex3y.dat')
y = np.expand_dims(y, axis=1)
#print x, y

m = len(y);

x
x = np.insert(x, 0, 1, 1)

x = np.matrix(x)
y = np.matrix(y)
mu = x.mean(0)
std = x.std(0)

print std[0,1]
#print mu[0,2]
#print type(mu)
#print x[:,1]

x[:,1] = (x[:,1] - mu[0,1]) / std[0,1]
x[:,2] = (x[:,2] - mu[0,2]) / std[0,2]

print np.size(x[1,:])
print np.zeros((1,2))
```

```python
# Gradient descent

alpha = [0.01, 0.03, 0.1, 0.3, 1, 1.3]
max_itr = 100

theta_gr_descent = np.matrix(np.zeros(np.size(x[1,:])))
#theta_gr_descent.transpose

ps = ['b', 'r', 'g', 'k', 'b--', 'r--']

for i in range(len(alpha)):
    theta = np.matrix(np.zeros(np.size(x[1,:])))
    theta = theta.transpose()
        
    J = np.matrix(np.zeros((max_itr, 1)))
    
    #print x.dot(theta)
    
    for iteration in range(max_itr):
        #print J[0]
        
        #print x.shape, y.shape, theta.shape
        f = x.dot(theta) - y
        #print f.dot(f.transpose())
        J[iteration] = f.transpose().dot(f) / 2*m
        
        # The gradient
        grad = x.transpose().dot(x.dot(theta) - y) / m
        
        theta = theta - (alpha[i] * grad)
        
    plt.plot(range(50), J[0:50], ps[i])
    plt.hold
        
print theta

plt.legend(['0.01','0.03','0.1', '0.3', '1', '1.3'])
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')

plt.show()
```
